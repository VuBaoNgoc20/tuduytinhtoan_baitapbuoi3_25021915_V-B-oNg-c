```python
#Bai11
a,b,c=map(int,input().split())
if a<0 or b<0 or c<0:
  print('Không phải tam giác')
else:
  if a+b>c and a+c>b and b+c>a:
    if a==b==c:
      print("Tam giác đều")
    elif a**2+b**2==c**2 or a**2 + c**2 ==b**2 or b**2 +c**2==a**2:
      print("Tam giác vuông")
    elif a==b!=c or a==c!=b or b==c!=a:
      print("Tam giác cân")
    else:
      print('Tam giác thường')
  else:
    print("không phải tam giác")
```

```python
#Bai16
n=float(input())
frac=n-int(n)
floor_val=int(n)
ceil_val=int(n)
if frac == 0:
  print(int(n),int(n),int(n))
else:
  if n<0:
    floor_val-=1
    if frac==-0.5:
      round_val=int(n)-1
    else:
      round_val=round(n)
    print(ceil_val,floor_val,round_val)
  elif n>0:
    ceil_val+=1
    if frac==0.5:
      round_val=int(n)+1
    else:
      round_val=round(n)
    print(ceil_val,floor_val,round_val)
```
