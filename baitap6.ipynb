```python
#W6A4
s=input()
pairs=s.split()
d={}
for item in pairs:
  key,val=item.split(':')
  if key not in d:
    d[key]=val
  else:
    s=list(d[key])
    s.append(val)
    d[key]=s
print(d)
```

```python

#W6A10
def binary_search(a,x,l,r):
  while l<=r:
    m=(l+r)//2
    if a[m]==x:
      return True
    elif a[m]> x:
      r=m-1
    else:
      l=m+1
  return False

a=list(map(int,input().split()))
k=int(input())
d=[]
for i in range(len(a)):
  if binary_search(a,k-a[i],i+1,len(a)-1):
    if (a[i],k-a[i]) not in d:
      d.append((a[i],k-a[i]))
d.sort(key=lambda x: x[0])
print(d)

```
