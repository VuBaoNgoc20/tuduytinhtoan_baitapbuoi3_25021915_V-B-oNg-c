```python
#Bài 6
from math import*
a,b,c=map(int,input('Nhập 3 cạnh tam giác:').split())
if a+b>c and a+c>b and c+b>a:
  C=(a+b+c)/2
  S=sqrt(C*(C-a)*(C-b)*(C-c))
  print('%.1f' % S)
else:
  print('Không phải 3 cạnh của tam giác')
```

```python
#9
n=int(input())
match n:
  case 0:
    print('Không')
  case 1:
    print('Một')
  case 2:
    print("Hai")
  case 3:
    print("Ba")
  case 4:
    print("Bốn")
  case 5:
    print("Năm")
  case 6:
    print("Sáu")
  case 7:
    print("Bảy")
  case 8:
    print("Tám")
  case 9:
    print("Chín")
  case _:
    print("INVALID")
```
